1. Generate a normal distribution (mean = 7, standard deviation = 2) , for three different sample sizes 100,1000,100,000. Design a python class to get the following items:

a) Show the density curve for all the three sample sizes (5 points)

b) Explain the findings for all three samples (5 points)

c) Convert the simulated data into a standard normal distribution, and plot for all three different sample sizes. (5 points)

## Answer 
(b)
Those samples have same mean and std values, thus the density curve looks alike.
Larger sample sizes will result smoothier distribution curves and higher density at the center values.

```python
from cmath import pi

class Question1:
    samplesize = [100, 1000, 100000]
    mean_value = 7
    std_value = 2

    def normal(self, miu, sigma, sample_size):
        """
        miu: mean
        sigma: standard deviation
        xrange: a list 
        """
        value_min = miu-4*sigma
        value_max = miu+4*sigma
        xrange = np.linspace(value_min, value_max, num=sample_size)
        zscore = (xrange-miu)/sigma
        y = (1/(sigma*np.sqrt(2*pi)))*np.exp(-0.5*(zscore**2))
        return xrange, y

    def normalplot(self):
        for i in self.samplesize:
            s = np.random.normal(self.mean_value, self.std_value, i)
            #xrange, s = self.normal(miu=self.mean_value, sigma=self.std_value, sample_size=i)
            #(a)
            sns.distplot(s, hist=False)
            # plt.plot(xrange, s)
            plt.ylabel('Density')
            plt.xlabel('Sample Value')
            plt.title(f'Sample size = {i}')
            plt.show()
            #(c) Convert into standard normal distribution and plot
            s_standard = (s-self.mean_value)/self.std_value
            sns.distplot(s_standard, hist=False)
            #plt.plot(xrange, s_standard)
            plt.ylabel('Density')
            plt.xlabel('Standard Sample Value')
            plt.title(f'Sample size = {i}')
            plt.show()

answer1 = Question1()
answer1.normalplot()
```

2. Generate a standard normal distribution for three different sample sizes 100, 5000, 50,000. Design a python class to get the following items:

a) Show the density curve for all the three sample sizes (5 points)

b) Explain the findings for all three samples (5 points)

## Answer
(b)
Most datapoints fall around 0, less datapoints fall out both sides.
Larger sample sizes will result smoothier density curves and more density around standard sample value = 0.

```python
class Question2:
    samplesize = [100, 5000, 50000]
    
    for i in samplesize:
        s = np.random.standard_normal(i)
        sns.distplot(s, hist=False)
        plt.ylabel('Density')
        plt.xlabel('Standard Sample Value')
        plt.title(f'Sample size = {i}')
        plt.show()
```

3. Binomial Distribution simulation. Generate a random number between 0 and 1. If that number is 0.5 or more, then event it as fake. Otherwise not a fake for 500, 5000, and 500,000 trails. 

Design a python class to get the following items:

a) Show the density curve for all the three sample sizes (5 points)

b) Explain the findings for all three samples (5 points)

## Answer
(b)
The highest density occured at about 0.5*trials. For example, 500 trials will give result in most success probability at around 250 success cases.

```python
class Question3:
    trials = [500, 5000, 500000]

    for i in trials:
        randlist = list(np.random.random_sample(size =i))
        count_fake = 0
        for j in randlist:
            if j >= 0.5:
                count_fake += 1
        plt.hist(randlist, 30)
        plt.ylabel('Random Number Probability Density')
        plt.xlabel('Random Number Value')
        plt.title(f'Sample size = {i}')
        plt.show()

    for i in trials:
        # n: number of trials
        # p: probability of getting fake for a single trial
        # size: repeated tests for the whole x trials.
        dist = np.random.binomial(n=i, p=0.5, size=1000)
        sns.distplot(dist, hist=False)
        plt.ylabel('Binomial Density')
        plt.xlabel('Binomial Fake Count')
        plt.title(f'Sample size = {i}')
        plt.show()
```
